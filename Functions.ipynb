## Default Arguments

```python
def employeeDetails(name='Unknown', address='Unknown Address', position='None', 
                    job_type='None', qualification='Unknown Qualification', job_description='None'):
    print(f'Employee Name: {name}')
    print(f'Employee Address: {address}')
    print(f'Employee Position: {position}')
    print(f'Employee Job Type: {job_type}')
    print(f'Employee Qualification: {qualification}')
    print(f'Employee Job Description: {job_description}')
```

## Keyword Arguments

```python
def employeeDetails(name='Unknown', address='Unknown Address', position='None', 
                    job_type='None', qualification='Unknown Qualification', job_description='None'):
    print(f'Employee Name: {name}')
    print(f'Employee Address: {address}')
    print(f'Employee Position: {position}')
    print(f'Employee Job Type: {job_type}')
    print(f'Employee Qualification: {qualification}')
    print(f'Employee Job Description: {job_description}')
```

## All Arguments

```python
# *args
def employeeDetailsArgs(name, age, address, qualification, *args):
    # print(args)
    for data in args:
        print(data)
    print(name)
    print(age)
```

```python
employeeDetailsArgs('Ram', 30, 'Kathmandu', 'Full-Time')
```

## All Keyword Arguments

```python
# **kwargs
def empKwArgs(**kwargs):
    print(kwargs)
    for key, val in kwargs.items():
        print(f'{key}: {val}')
```

```python
empKwArgs(Name='Ram', Age=30, Address='Kathmandu', Job_Type='Full-Time')
```

## Nested Function

```python
def add(a, b):
    return f'Addition of {a} and {b} is: {a + b}.'

def sub(a, b):
    return f'Subtraction of {a} and {b} is {a - b}.'

def mul(a, b):
    return f'Multiplication of {a} and {b} is {a * b}.'

def div(a, b):
    return f'Division of {a} and {b} is {a/b}.'
```

```python
def calculator():
    choice = int(input('Enter your choice: '))
    num1 = float(input('Enter first number: '))
    num2 = float(input('Enter your second number: '))
    if choice == 1:
        print(add(num1, num2))
    elif choice == 2:
        print(sub(num1, num2))
    elif choice == 3:
        print(mul(num1, num2))
    elif choice == 4:
        print(div(num1, num2))
    else:
        print('Invalid Input')
```

```python
calculator()
```

## Lambda Functions

```python
addition = lambda a, b : f'Addition of {a} and {b} is: {a + b}'
```

```python
addition(5, 6)
```

```python
odd_even = lambda : [f"Even: {i}" if i % 2 == 0 else f"Odd: {i}" for i in range(20)]
```

```python
# odd_even()
```

```python
prices = [100, 250, 75, 400, 120, 310]
```

```python
get_cheap = lambda lst : [price for price in lst if price < 200]

get_expensive = lambda lst : [price for price in lst if price >= 200]
```

```python
get_cheap(prices)
```

```python
get_expensive(prices)
```

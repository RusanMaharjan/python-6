```python
'''
Grade Calculation
--------------
marks 80 - 100 -> Distinction
marks 70 - 79 -> A
marks 60 - 69 -> B
marks 50 - 59 -> C
marks 40 - 49 -> D
marks < 40 -> Fail
'''
```

```python
marks = float(input('Enter a marks: '))

if marks >= 80 and marks <= 100:
    print(f'Marks: {marks}\nGrade - Distinction')
elif marks >= 70 and marks < 80:
    print(f'Marks: {marks}\nGrade - A')
elif marks >= 60 and marks < 70:
    print(f'Marks: {marks}\nGrade - B')
elif marks >= 50 and marks < 60:
    print(f'Marks: {marks}\nGrade - C')
elif marks >= 40 and marks < 50:
    print(f'Marks: {marks}\nGrade - D')
elif marks >=0 and marks < 40: # 0 - 39
    print(f'Marks:{marks}\nGrade - Fail')
else:
    print(f'{marks}: Invalid Marks!! Please input valid marks (0 - 100).')
```

```python
'''
both username and password -> Login successful.
username invalid and password valid -> user not found.
username valid and password invalid -> incorrect password. Try again.
'''
```

```python
uname = 'abc@gmail.com'
pwd = '1234567890'

username = input('Enter your username: ')
password = input('Enter your password: ')

if username == uname and password == pwd:
    print('Login successful.')
elif username != uname and password == pwd:
    print('User not found!!')
elif username == uname and password != pwd:
    print('Incorrect Password!!')
else:
    print('Invalid credintials. Try again')
```

```python
'''
tax based on annual income 
<=250000 then no tax 
250001 - 500000 then 5% tax 
500001 - 1000000 then 10% 
more than 1000000 then 20% 

output
-------------
Annual Income: Rs.400000
Taxable Income(5%): Rs.20000
Remaining Income = Rs.380000
'''
```

```python
annual_income = int(input('Enter your annual income: '))
remaining_income = 0
taxable_income = 0

# 0 - 250000
if annual_income > 0 and annual_income <= 250000:
    print(f'Annual Income: Rs.{annual_income}.')
    print(f'Taxable Income (0%): Rs.{taxable_income}.')
    print(f'Remaining Income: Rs.{annual_income}')

# 250001 - 500000
elif annual_income > 250000 and annual_income <= 500000:
    print(f'Annual Income: Rs.{annual_income}.')
    taxable_income = annual_income * 0.05
    print(f'Taxable Income (5%): Rs.{taxable_income}')
    remaining_income = annual_income - taxable_income # annual_income -= taxable_income
    print(f'Remaining Income: Rs.{remaining_income}')

# 500001 - 1000000
elif annual_income > 500000 and annual_income <= 1000000:
    print(f'Annual Income: Rs.{annual_income}.')
    taxable_income = annual_income * 0.10
    print(f'Taxable Income (10%): Rs.{taxable_income}')
    remaining_income = annual_income - taxable_income # annual_income -= taxable_income
    print(f'Remaining Income: Rs.{remaining_income}')

# more than 1000000
elif annual_income > 1000000:
    print(f'Annual Income: Rs.{annual_income}.')
    taxable_income = annual_income * 0.20
    print(f'Taxable Income (20%): Rs.{taxable_income}')
    remaining_income = annual_income - taxable_income # annual_income -= taxable_income
    print(f'Remaining Income: Rs.{remaining_income}')
else:
    print('Invalid Annual Income.')
```

```python
## Balance withdrawl.
# pin number , current balance, withdraw, 1000s
# Incorrect pin
# Withdrawl amount must be multiple of 1000.

pin = 1234
current_balance = 50000

user_pin = int(input('Enter your pin number: '))
balance = int(input('Enter your withdraw balance: '))

if user_pin == pin and balance < current_balance and balance % 1000 == 0: # True False False = False
    print(f'Rs.{balance} has been withdrawn from A/C no. 1234567890.')
elif user_pin != pin:
    print('Incorrect Pin!')
elif balance > current_balance:
    print('Balance amount cannot be more than your initial balance.')
elif  balance % 1000 != 0:
    print('Withdrawl amount must be multiple of 1000.')
else:
    print('Invalid Input..')
```

```python
# Student Exam Eligiblity
# attendance >= 75
# internal_marks >= 40
attendance = 50
internal_marks = 60

if attendance >= 75 and internal_marks >= 40:
    print('You are eligible to give examination.')
elif attendance < 75:
    print('Attendance requirement does not meet. Not eligible for examination.')
elif internal_marks < 40:
    print('Internal marks requirement does not meet. Not eligible for examination.')
else:
    print('Invalid input..')
```

## Match Case

```python
# 1. Breakfast
# 2. Lunch
# 3. Snacks
# 4. Dinner
# 5. Exit

print('1. Breakfast')
print('2. Lunch')
print('3. Snacks')
print('4. Dinner')
print('5. Exit')

choice = input('Enter your order choice (1 - 5): ')

match choice:
    case '1':
        print('You have ordered Breakfast.')
    case '2':
        print('You have ordered Lunch.')
    case '3':
        print('You have ordered Snacks.')
    case '4':
        print('You have ordered Dinner.')
    case '5':
        print('Thank you for choosing us.')
    case _: # else
        print('Invalid number.')
```

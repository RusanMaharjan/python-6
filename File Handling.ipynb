## CSV Data (Comma Separated Value)

```python
import csv
```

```python
country_destination = [
    ['Country', 'Destination', 'Price', 'Days to Stay'],
    ['Italy', 'Rome', '$5000', 10],
    ['Nepal', 'Kathmandu', '$3000', 5],
    ['Canada', 'Ottawa', '$4000', 10]
]
```

```python
# open(file_name, operation, newline='')
with open('country.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(country_destination)
```

```python
## Read CSV File
with open('country.csv', 'r') as file:
    reader = csv.reader(file)
    for data in reader:
        print(data)
```

```python
new_country_destination = [
    ['India', 'New Delhi', '$5000', 6],
    ['USA', 'New York', '$9000', 7]
]
```

```python
## append data in csv
with open('country.csv', 'a', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(new_country_destination)
```

### Dictionary Data to CSV

```python
products = [
    {
        'Product_Name': 'Mobile',
        'Product_Desc': 'Black Mobile',
        'Price': 30000,
        'Quantity': 3
    },
    {
        'Product_Name': 'Bottle',
        'Product_Desc': 'White Bottle',
        'Price': 200,
        'Quantity': 4
    },
    {
        'Product_Name': 'Charger',
        'Product_Desc': 'Green Mobile',
        'Price': 400,
        'Quantity': 5
    }
]
```

```python
## Create file
with open('products.csv', 'w', newline='') as file:
    columns = ['Product_Name', 'Product_Desc', 'Price', 'Quantity']
    writer = csv.DictWriter(file, fieldnames = columns)
    writer.writeheader()
    writer.writerows(products)
```

```python
## Read Dictionary Data
with open('products.csv', 'r') as file:
    reader = csv.DictReader(file)
    for data in reader:
        for key, val in data.items():
            print(f'{key}: {val}')
        print('-'*20)
```

```python
new_products = [
    {
        'Product_Name': 'Bag',
        'Product_Desc': 'Black Bag',
        'Price': 3000,
        'Quantity': 2
    },
    {
        'Product_Name': 'TV',
        'Product_Desc': 'MI TV',
        'Price': 150000,
        'Quantity': 1
    }
]
```

```python
with open('products.csv', 'a', newline='') as file:
    columns = ['Product_Name', 'Product_Desc', 'Price', 'Quantity']
    writer = csv.DictWriter(file, fieldnames=columns)
    writer.writerows(new_products)
```

## JSON (Javascript Object Notation)

```python
products = [
    {
        'Product_Name': 'Mobile',
        'Product_Desc': 'Black Mobile',
        'Price': 30000,
        'Quantity': 3
    },
    {
        'Product_Name': 'Bottle',
        'Product_Desc': 'White Bottle',
        'Price': 200,
        'Quantity': 4
    },
    {
        'Product_Name': 'Charger',
        'Product_Desc': 'Green Mobile',
        'Price': 400,
        'Quantity': 5
    }
]
```

```python
'''
file create -> dump()
file read -> load()
'''
```

```python
import json
```

```python
# Create json file
# open(file_name, operation)
with open('products.json', 'w') as file:
    json.dump(products, file, indent=4)
```

```python
# read json file
with open('products.json', 'r') as file:
    data = json.load(file)
    for row in data:
        for key, val in row.items():
            print(f'{key}: {val}')
        print('-'*20)
```

```python
new_data = {
    'Product_Name': 'Bag',
    'Product_Desc': 'Black Bag',
    'Price': 3000,
    'Quantity': 2
}
```

```python
# append data in json
with open('products.json', 'w') as file:
    data.append(new_data)
    json.dump(data, file, indent=4)
```

```python
# num1 = int(input('Enter first number: '))
# num2 = int(input('Enter second numberr: '))
try:
    num1 = int(input('Enter first number: '))
    num2 = int(input('Enter second number: '))
    num3 = num1 / num2
    print(num3)
except ZeroDivisionError:
    print('Error: Number cannot be divisible by zero.')
except ValueError:
    print('Error: Only integer values are accepted in input field (0 - 9).')
# except ValueError as abc:
#     print(abc)
```

```python
try:
    file_name = input('Enter your text file name: ')
    with open(f'{file_name}.txt', 'r') as file:
        data = file.read()
except FileNotFoundError:
    print(f'Error: No such file found in folder named: "{file_name}.txt"')
else:
    print(data)
finally:
    print('Program ended successfully.')
```

### Custom Exception

```python
class AmountWithdrawError(Exception):
    pass
```

```python
def withdrawAmount():
    balance = 50000
    withdraw_amount = int(input('Enter your withdraw balance: '))
    
    if withdraw_amount < balance:
        balance -= withdraw_amount
        print(f'Rs.{withdraw_amount} successfully withdrawn from account.')
    else:
        raise AmountWithdrawError('Amount must be less than actual balance.')
```

```python
try:
    withdrawAmount()
except AmountWithdrawError as e:
    print(e)
```

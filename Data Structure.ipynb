```python
# Lab Question 5: The "Price Is Right" (Conditional Logic) You are given a list of product 
# prices: prices = [100, 250, 75, 400, 120, 310]. 
# 1. Write a program that creates two new empty lists: cheap_items and 
# expensive_items. 
# 2. Loop through the prices list. If a price is greater than 200, append it to 
# expensive_items. Otherwise, append it to cheap_items. 
# 3. Print both lists and calculate the total count of items in each using the len() 
# function.
```

```python
prices = [100, 250, 75, 400, 120, 310]
cheap_items = []
expensive_items = []
```

```python
for price in prices:
    if price >= 200:
        expensive_items.append(price)
    else:
        cheap_items.append(price)
```

```python
print(expensive_items)
print(cheap_items)
```

```python
print(f'Total Expensive Items: {len(expensive_items)} items')
print(f'Total Cheap Items: {len(cheap_items)} items')
```

```python
# Lab Question 6: Matrix Modification (2D Lists) Given this 3x3 matrix: matrix = [[1, 2, 3], [4, 
# 5, 6], [7, 8, 9]] 
# 1. Change the middle value (5) to 0 using its row and column index. 
# 2. Add a new row [10, 11, 12] to the end of the matrix. 
# 3. Access the last element of the second row (6) and the first element of the fourth row 
# (10), and print their sum. 
```

```python
matrix = [
            [1, 2, 3], 
            [4, 5, 6],
            [7, 8, 9]
]
matrix
```

```python
for row in matrix[1:2]:
    row[1] = 0
matrix
```

```python
matrix.append([10, 11, 12])
matrix
```

```python
for row in matrix[1:2]:
    last_val = row[2]

for row in matrix[3:]:
    first_val =  row[0]

print(first_val)
print(last_val)
```

```python
last_val + first_val
```

## Dictionary

```python
# {}
data = {
    'Product_Name': 'Mobile',
    'Product_Desc': 'Black Mobile',
    'Price': 30000,
    'Quantity': 3
}
data
```

```python
prod_data = {
    'Product_Name': ['Mobile', 'Bottle', 'Charger'],
    'Product_Desc': ['Black Mobile', 'White Bottle', 'Green Charger'],
    'Price': [30000, 200, 400],
    'Quantity': [3, 4, 5]
}
prod_data
```

```python
# JSON (Javascript Object Notation)
prod_json = [
    {
        'Product_Name': 'Mobile',
        'Product_Desc': 'Black Mobile',
        'Price': 30000,
        'Quantity': 3
    },
    {
        'Product_Name': 'Bottle',
        'Product_Desc': 'White Bottle',
        'Price': 200,
        'Quantity': 4
    },
    {
        'Product_Name': 'Charger',
        'Product_Desc': 'Green Mobile',
        'Price': 400,
        'Quantity': 5
    }
]
```

```python
'''
- .keys() -> Used to print keys of dictionary
- .values() -> Used to print values of dictionary
- .items() -> Used to print both key and values of dictionary
'''
```

```python
for key in data.keys():
    print(key)
```

```python
for value in data.values():
    print(value)
```

```python
for key, value in data.items():
    print(f'{key}: {value}')
```

```python
for key, val in prod_data.items():
    print(f'{key}: {val}\n')
    for data in val:
        print(f'{key}: {data}')
    print('-'*55)
```

## Inheritance

```python
'''
# Simple Inheritance -> two classes -> parent class and child class
# Multiple Inheritance -> Hierarchical Inheritance -> Generalization Hierarchy
    -> 3 or more than 3 classes -> one child inherits multiple parents

# Multi-Level Inhertiance
    -> 3 or more than 3 classes -> Grand Parent, Parent and Child
'''
```

## Simple Inheritance

```python
# Parent class
class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone):
        self.customer_id = customer_id
        self.customer_name = customer_name
        self.customer_address = customer_address
        self.customer_phone = customer_phone

    def customer_details(self):
        print('Customer Details')
        print('-'*55)
        print(f'Customer Id: {self.customer_id}')
        print(f'Customer Name: {self.customer_name}')
        print(f'Customer Address: {self.customer_address}')
        print(f'Customer Phone: {self.customer_phone}')
```

```python
# Child Class
class Product(Customer):
    def __init__(self, product_id, product_name, product_quantity, list_price, product_description,
                customer_id, customer_name, customer_address, customer_phone):
        # Super Keyword is used to inherit parent class constructor.
        super().__init__(customer_id, customer_name, customer_address, customer_phone)
        self.product_id = product_id
        self.product_name = product_name
        self.product_quantity = product_quantity
        self.list_price = list_price
        self.product_description = product_description
        self.total_price = product_quantity * list_price

    def product_details(self):
        print('\nProduct Details')
        print('-'*55)
        print(f'Product Id: {self.product_id}')
        print(f'Product Name: {self.product_name}')
        print(f'Product Quantity: {self.product_quantity}')
        print(f'Product List Price: Rs.{self.list_price}')
        print(f'Product Description: {self.product_description}')
        print(f'Total Price: Rs.{self.total_price}')
```

```python
p = Product(100, 'Bag', 3, 4000, 'Leather Black Bag',
            'Cust 1', 'Bob', 'Jhapa', 1234567890)
```

```python
p.customer_details()
p.product_details()
```

## Multiple Inheritance

```python
# Parent class
class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone):
        self.customer_id = customer_id
        self.customer_name = customer_name
        self.customer_address = customer_address
        self.customer_phone = customer_phone

    def customer_details(self):
        print('Customer Details')
        print('-'*55)
        print(f'Customer Id: {self.customer_id}')
        print(f'Customer Name: {self.customer_name}')
        print(f'Customer Address: {self.customer_address}')
        print(f'Customer Phone: {self.customer_phone}')
```

```python
class Staff:
    def __init__(self, staff_id, staff_name, staff_position, staff_job_type):
        self.staff_id = staff_id
        self.staff_name = staff_name
        self.staff_position = staff_position
        self.staff_job_type = staff_job_type

    def staff_details(self):
        print(f'Staff Id: {self.staff_id}')
        print(f'Staff Name: {self.staff_name}')
        print(f'Staff Position: {self.staff_position}')
        print(f'Staff Job Type: {self.staff_job_type}')
```

```python
class Product(Customer, Staff):
    def __init__(self, product_id, product_name, product_quantity, list_price, product_description,
                customer_id, customer_name, customer_address, customer_phone,
                staff_id, staff_name, staff_position, staff_job_type):
        # Constructor Inheritance
        Customer.__init__(self, customer_id, customer_name, customer_address, customer_phone)
        Staff.__init__(self, staff_id, staff_name, staff_position, staff_job_type)
        
        self.product_id = product_id
        self.product_name = product_name
        self.product_quantity = product_quantity
        self.list_price = list_price
        self.product_description = product_description
        self.total_price = product_quantity * list_price

    def product_details(self):
        print('\nProduct Details')
        print('-'*55)
        print(f'Product Id: {self.product_id}')
        print(f'Product Name: {self.product_name}')
        print(f'Product Quantity: {self.product_quantity}')
        print(f'Product List Price: Rs.{self.list_price}')
        print(f'Product Description: {self.product_description}')
        print(f'Total Price: Rs.{self.total_price}')
```

```python
p = Product(100, 'Bag', 3, 4000, 'Leather Black Bag',
            'Cust 1', 'Bob', 'Jhapa', 1234567890,
            'St-1', 'Bill', 'Junior', 'Full-Time')
```

```python
p.staff_details()
```

```python
p.customer_details()
```

```python
p.product_details()
```

```python
# Parent class
class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone, **kwargs):
        super().__init__(**kwargs)
        self.customer_id = customer_id
        self.customer_name = customer_name
        self.customer_address = customer_address
        self.customer_phone = customer_phone

    def customer_details(self):
        print('Customer Details')
        print('-'*55)
        print(f'Customer Id: {self.customer_id}')
        print(f'Customer Name: {self.customer_name}')
        print(f'Customer Address: {self.customer_address}')
        print(f'Customer Phone: {self.customer_phone}')
```

```python
class Staff:
    def __init__(self, staff_id, staff_name, staff_position, staff_job_type, **kwargs):
        super().__init__(**kwargs)
        self.staff_id = staff_id
        self.staff_name = staff_name
        self.staff_position = staff_position
        self.staff_job_type = staff_job_type

    def staff_details(self):
        print(f'Staff Id: {self.staff_id}')
        print(f'Staff Name: {self.staff_name}')
        print(f'Staff Position: {self.staff_position}')
        print(f'Staff Job Type: {self.staff_job_type}')
```

```python
class Product(Customer, Staff):
    def __init__(self, product_id, product_name, product_quantity, list_price, product_description,
                customer_id, customer_name, customer_address, customer_phone,
                staff_id, staff_name, staff_position, staff_job_type):
        # Constructor Inheritance
        super().__init__(customer_id=customer_id, customer_name=customer_name, customer_address=customer_address, 
                         customer_phone=customer_phone, 
                         staff_id=staff_id, staff_name=staff_name, 
                        staff_position=staff_position, staff_job_type=staff_job_type)
        
        self.product_id = product_id
        self.product_name = product_name
        self.product_quantity = product_quantity
        self.list_price = list_price
        self.product_description = product_description
        self.total_price = product_quantity * list_price

    def product_details(self):
        print('\nProduct Details')
        print('-'*55)
        print(f'Product Id: {self.product_id}')
        print(f'Product Name: {self.product_name}')
        print(f'Product Quantity: {self.product_quantity}')
        print(f'Product List Price: Rs.{self.list_price}')
        print(f'Product Description: {self.product_description}')
        print(f'Total Price: Rs.{self.total_price}')
```

```python
pa = Product(100, 'Bag', 3, 4000, 'Leather Black Bag',
            'Cust 1', 'Bob', 'Jhapa', 1234567890,
            'St-1', 'Bill', 'Junior', 'Full-Time')
```

## Multi Level Inheritance

```python
# Customer -> Product -> Order
# Parent class
class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone):
        self.customer_id = customer_id
        self.customer_name = customer_name
        self.customer_address = customer_address
        self.customer_phone = customer_phone

    def customer_details(self):
        print('Customer Details')
        print('-'*55)
        print(f'Customer Id: {self.customer_id}')
        print(f'Customer Name: {self.customer_name}')
        print(f'Customer Address: {self.customer_address}')
        print(f'Customer Phone: {self.customer_phone}')
```

```python
class Product(Customer):
    def __init__(self, product_id, product_name, product_quantity, list_price, product_description,
                customer_id, customer_name, customer_address, customer_phone):
        # Super Keyword is used to inherit parent class constructor.
        super().__init__(customer_id, customer_name, customer_address, customer_phone)
        self.product_id = product_id
        self.product_name = product_name
        self.product_quantity = product_quantity
        self.list_price = list_price
        self.product_description = product_description
        self.total_price = product_quantity * list_price

    def product_details(self):
        print('\nProduct Details')
        print('-'*55)
        print(f'Product Id: {self.product_id}')
        print(f'Product Name: {self.product_name}')
        print(f'Product Quantity: {self.product_quantity}')
        print(f'Product List Price: Rs.{self.list_price}')
        print(f'Product Description: {self.product_description}')
        print(f'Total Price: Rs.{self.total_price}')
```

```python
class Order(Product):
    def __init__(self, order_id, order_desc, order_destination, order_quantity,
                product_id, product_name, product_quantity, list_price, product_description,
                customer_id, customer_name, customer_address, customer_phone):
        
        super().__init__(product_id, product_name, product_quantity, list_price, product_description,
                customer_id, customer_name, customer_address, customer_phone)
        self.order_id = order_id
        self.order_desc = order_desc
        self.order_destination = order_destination
        self.order_quantity = order_quantity

    def order_details(self):
        print('Order Details')
        print('-'*55)
        print(f'Order Id: {self.order_id}')
        print(f'Order Description: {self.order_desc}')
        print(f'Order Destination: {self.order_destination}')
        print(f'Order Quantity: {self.order_quantity}')

        print('\nCustomer Details')
        print('-'*55)
        print(f'Customer Id: {self.customer_id}')
        print(f'Customer Name: {self.customer_name}')
        print(f'Customer Address: {self.customer_address}')
        print(f'Customer Phone: {self.customer_phone}')

        print('\nProduct Details')
        print('-'*55)
        print(f'Product Id: {self.product_id}')
        print(f'Product Name: {self.product_name}')
        print(f'Product Quantity: {self.product_quantity}')
        print(f'Product List Price: Rs.{self.list_price}')
        print(f'Product Description: {self.product_description}')
        print(f'Total Price: Rs.{self.total_price}')
```

```python
o = Order(10, 'Mobile package', 'Lalitpur', 2,
          101, 'POCO Phone', 2, 40000, 'Red POCO Phone',
          'CUS19', 'Ram', 'Lalitpur', 1234567890)

o.order_details()
```

```python
o.product_details()
```

```python
o.customer_details()
```

# Encapsulation

```python
'''
Access Modifiers
-------------------
Private (__) -> Value access only inside the class
Public
Protected (_) -> Can be directly accessable inside and outside the class.

self.__order_id (Private)
self._order_id (Protected)
'''
```

```python
class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone):
        self.__customer_id = customer_id
        self.__customer_name = customer_name
        self.__customer_address = customer_address
        self.__customer_phone = customer_phone

    def customer_details(self):
        print('Customer Details')
        print('-'*55)
        print(f'Customer Id: {self.__customer_id}')
        print(f'Customer Name: {self.__customer_name}')
        print(f'Customer Address: {self.__customer_address}')
        print(f'Customer Phone: {self.__customer_phone}')
```

```python
class Product(Customer):
    def __init__(self, product_id, product_name, product_quantity, list_price, product_description,
                customer_id, customer_name, customer_address, customer_phone):
        # Super Keyword is used to inherit parent class constructor.
        super().__init__(customer_id, customer_name, customer_address, customer_phone)
        self.__product_id = product_id
        self.__product_name = product_name
        self.__product_quantity = product_quantity
        self.__list_price = list_price
        self.__product_description = product_description
        self.__total_price = product_quantity * list_price

    def product_details(self):
        print('\nProduct Details')
        print('-'*55)
        print(f'Product Id: {self.__product_id}')
        print(f'Product Name: {self.__product_name}')
        print(f'Product Quantity: {self.__product_quantity}')
        print(f'Product List Price: Rs.{self.__list_price}')
        print(f'Product Description: {self.__product_description}')
        print(f'Total Price: Rs.{self.__total_price}')
```

```python
p = Product(101, 'POCO Phone', 2, 40000, 'Red POCO Phone',
          'CUS19', 'Ram', 'Lalitpur', 1234567890)
```

```python
p.customer_details()
p.product_details()
```

```python
class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone):
        self._customer_id = customer_id
        self._customer_name = customer_name
        self._customer_address = customer_address
        self._customer_phone = customer_phone

    def __customer_details(self):
        print('Customer Details')
        print('-'*55)
        print(f'Customer Id: {self._customer_id}')
        print(f'Customer Name: {self._customer_name}')
        print(f'Customer Address: {self._customer_address}')
        print(f'Customer Phone: {self._customer_phone}')

    def c_details(self):
        self.__customer_details()
```

```python
class Product(Customer):
    def __init__(self, product_id, product_name, product_quantity, list_price, product_description,
                customer_id, customer_name, customer_address, customer_phone):
        # Super Keyword is used to inherit parent class constructor.
        super().__init__(customer_id, customer_name, customer_address, customer_phone)
        self._product_id = product_id
        self._product_name = product_name
        self._product_quantity = product_quantity
        self._list_price = list_price
        self._product_description = product_description
        self._total_price = product_quantity * list_price

    def product_details(self):
        print('\nProduct Details')
        print('-'*55)
        print(f'Product Id: {self._product_id}')
        print(f'Product Name: {self._product_name}')
        print(f'Product Quantity: {self._product_quantity}')
        print(f'Product List Price: Rs.{self._list_price}')
        print(f'Product Description: {self._product_description}')
        print(f'Total Price: Rs.{self._total_price}')
```

```python
p = Product(101, 'POCO Phone', 2, 40000, 'Red POCO Phone',
          'CUS19', 'Ram', 'Lalitpur', 1234567890)
```

```python
p.c_details()
p.product_details()
```

## Polymorphism

```python
'''
Polymorphism
-------------
One function multiple similar work.

Poly -> Many
Morph -> Forms
'''
```

```python
class Viber:
    def message(self, send_from, send_to, message):
        print(f'{send_to}: Viber Message')
        print(f'{send_from} sent you a message.')
        print(f'Message: {message}')

class WhatsApp:
    def message(self, send_from, send_to, message):
        print(f'{send_to}: WhatsApp Message')
        print(f'{send_from} sent you a message.')
        print(f'Message: {message}')

class Messenger:
    def message(self, send_from, send_to, message):
        print(f'{send_to}: Messenger Message')
        print(f'{send_from} sent you a message.')
        print(f'Message: {message}')

class Slack:
    def message(self, send_from, send_to, message):
        print(f'{send_to}: Slack Message')
        print(f'{send_from} sent you a message.')
        print(f'Message: {message}')
```

```python
def send_message(message_obj, send_from, send_to, message):
    message_obj.message(send_from, send_to, message)
```

```python
send_message(WhatsApp(), 'Bob', 'Ram', 'Hello.')
```

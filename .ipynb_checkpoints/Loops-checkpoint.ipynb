```python
# Write a program to print 10 natural number.
# for(i=0; i<=10;i+2){}
for i in range(1, 11): # 11 < 11 -> False
    print(i)
```

```python
# Write a program to check number is odd or even of 20 natural number.
for i in range(1, 21):
    if i % 2 == 0:
        print(f'{i}: Even Number.')
    else:
        print(f'{i}: Odd Number')
```

```python
# Multiplication Table
# 3 x 1 = 3
# 3 x 2 = 6
# ...
# 3 x 10 = 30
num = int(input('Enter a number: '))
for i in range(1, 11):
    print(f'{num} x {i} = {num * i}')
```

```python
for i in range(3): # 0, 1, 2
    name = input('\nEnter name of student: ')
    marks = float(input('Enter a marks: '))

    print(f'\nName of student: {name}')
    if marks >= 80 and marks <= 100:
        print(f'Marks: {marks}\nGrade - Distinction')
    elif marks >= 70 and marks < 80:
        print(f'Marks: {marks}\nGrade - A')
    elif marks >= 60 and marks < 70:
        print(f'Marks: {marks}\nGrade - B')
    elif marks >= 50 and marks < 60:
        print(f'Marks: {marks}\nGrade - C')
    elif marks >= 40 and marks < 50:
        print(f'Marks: {marks}\nGrade - D')
    elif marks >=0 and marks < 40: # 0 - 39
        print(f'Marks:{marks}\nGrade - Fail')
    else:
        print(f'{marks}: Invalid Marks!! Please input valid marks (0 - 100).')
```

```python
for i in range(5):
    print('1. Breakfast')
    print('2. Lunch')
    print('3. Snacks')
    print('4. Dinner')
    print('5. Exit')
    
    choice = input('Enter your order choice (1 - 5): ')
    
    match choice:
        case '1':
            print('You have ordered Breakfast.')
        case '2':
            print('You have ordered Lunch.')
        case '3':
            print('You have ordered Snacks.')
        case '4':
            print('You have ordered Dinner.')
        case '5':
            print('Thank you for choosing us.')
            break
        case _: # else
            print('Invalid number.')
else:
    print('Order limit.')
```

## Break and Continue

```python
'''
- Break -> To stop program forcefully.
- Continue -> To skip values/condition.
'''
```

```python
for i in range(1, 11):
    print(i)
    if i == 5:
        break
```

```python
for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```

```python
# ATM Card Blocked. 
pin = 1234
current_balance = 50000
remaining_attempt = 3

for i in range(3):
    print(f'\nRemaining Attempt: {remaining_attempt}')
    print('-'*30)
    user_pin = int(input('Enter your pin number: '))
    
    if user_pin == pin:
        balance = int(input('Enter your withdraw balance amount: '))
        if balance < current_balance:
            if balance % 1000 == 0:
                # withdraw_balance = current_balance - balnce
                current_balance -= balance
                print(f'Rs.{balance} has been withdrawn from A/C no. 1234567890987654.')
                print(f'Remaining Balance: Rs.{current_balance}')
                break
            else:
                print('Withdraw balance must be multiple of 1000s.')
        else:
            print('Withdraw balance must be less than your current balance.')
    else:
        print('Incorrect Pin Number. Try Again.')
        remaining_attempt -= 1
else:
    print('\nCard Blocked. Please contact nearest branch.')
```

## While Loop

```python
'''
while condition:
    logic..
    print..

- While loop runs until condition becomes False.
'''
```

```python
secret_value = 7
guess_value = 0

while guess_value != secret_value: # 0 != 7 -> False -> break
    guess_value = int(input('Enter a guess number: '))

    if guess_value == secret_value: # 7 == 7 -> True
        print(f'Guessed Correct Value: {guess_value}')
    else:
        print('Incorrect Value. Try Again!!')
```

```python
secret_value = 7
guess_value = 0

while True:
    guess_value = int(input('Enter a number: '))

    if guess_value == secret_value: # 7 == 7 -> True
        print('Correct Value.')
        break
    else:
        print('Incorrect Value. Try Again.')
```

```python
choice = 0

while choice != 5:
    print('1. Momo') 
    print('2. Chowmin') 
    print('3. Pizza') 
    print('4. Drinks') 
    print('5. Exit')

    choice = int(input('\nEnter your choice (1 - 5): '))

    if choice == 1:
        print('You have ordered Momo.')
    elif choice == 2:
        print('You have ordered Chowmin.')
    elif choice == 3:
        print('You have ordered Pizza.')
    elif choice == 4:
        print('You have ordered Drinks.')
    elif choice == 5:
        print('Thank you for choosing us.')
    else:
        print('Invalid Input.')
```

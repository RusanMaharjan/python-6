# Operators

```python
'''
1. Arithmetic Operator
  -> +, -, *, /, % (modulus / Reminder), // (Float Division)
  
2. Logical Operator
  -> And, Or, Not
  
3. Comparison Operator
  -> =, ==, >, >=, <, <=

  True / False
'''
```

```python
num_a = int(input('Enter first number: '))
num_b = int(input('Enter second number: '))

add = num_a + num_b
sub = num_a - num_b
mul = num_a * num_b
div = num_a / num_b
rem = num_a % num_b
fd = num_a // num_b

print(f'Addition of num1 and num2: {add}')
print(f'Subtraction of num1 and num2: {sub}')
print(f'Multiplication of num1 and num2: {mul}')
print(f'Division of num1 and num2: {div}')
print(f'Reminder of num1 and num2: {rem}')
print(f'Float Division of num1 and num2: {fd}')
```
